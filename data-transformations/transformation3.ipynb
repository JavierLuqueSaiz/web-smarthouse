```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data3.csv")
```

```python
df = df.drop(["Unnamed: 0.1", "Unnamed: 0"], axis=1)
```

```python
df.head()
```

```python
df["Mes"] = df.time.str[5:7].astype(int)
df.head()
```

```python
df["Dia"] = df.time.str[8:10].astype(int)
df.head()
```

```python
df["Hora"] = df.time.str[11:13].astype(int)
df.head()
```

```python
df["Minuto"] = df.time.str[14:16].astype(int)
df.head()
```

```python
df = pd.get_dummies(df, columns=["Mes","Hora"])
```

```python
df.to_csv("data4.csv")
```
